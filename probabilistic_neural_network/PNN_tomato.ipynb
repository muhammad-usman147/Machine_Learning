```python
import numpy as np 
from sklearn import datasets, metrics 

import os
from sklearn.model_selection import train_test_split
```

## Data Preprocessing

```python
os.listdir('/content/DATASET/tomato')
```

```python
DIR = "/content/DATASET/tomato/"
for name in os.listdir(DIR):
  print("Class name: {} , Total Samples: {}".format(name, 
                                                     len(os.listdir(DIR+name))))
```

```python
import cv2
all_X = []
all_Y = []
DIR = "/content/DATASET/tomato/"
for name in os.listdir(DIR):
  print("Loading images for : {}".format(name))
  for img in os.listdir(DIR+name):
    all_X.append(cv2.imread(DIR+name+"/"+img,0))
    if name == 'Tomato_healthy':
      all_Y.append(0)
    elif name == 'Tomato__Target_Spot':
      all_Y.append(1)
    elif name == 'Tomato__Tomato_mosaic_virus':
      all_Y.append(2)
    elif name == 'Tomato_Early_blight':
      all_Y.append(3)
    elif name == 'Tomato_Bacterial_spot':
      all_Y.append(4)
```

```python
X = np.array(all_X)
Y = np.array(all_Y)
X.shape, Y.shape
```

```python
X = np.array(all_X)
Y = np.array(all_Y)
X.shape, Y.shape
```

```python
print("Images value min: {} max: {} mean: {} ".format(X.min(),
                                                      X.max(),
                                                      X.mean()) )
```

```python
print("Reshaping the dimensions of the images into ( {}, {}*{})".format(X.shape[0],X.shape[1],X.shape[2]))
X_ =np.reshape(X,(X.shape[0],X.shape[1]*X.shape[2]))
X_.shape
```

# Training The Model Without normalization

```python
print("Splitting the data")
xtrain, xtest, ytrain, ytest = train_test_split(X_,Y, test_size = 0.4,random_state = 100)
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
from neupy import algorithms 

model = algorithms.PNN(std=xtrain.mean(), batch_size=200, verbose=True)
model.train(xtrain,ytrain)
```

## Testing ACC and Confusion Matrix

### accuracy score

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(xtest)
accuracy_score(ytest,y_pred)
```

### confusion matrix

```python
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
confusion_matrix(ytest,y_pred)
```
